```python
ukr = pd.read_csv('data/COUNT_NATIONAL_IN_HOTEL/Turkey/ukrInTurkeyHotelsCount.csv')
rus = pd.read_csv('data/COUNT_NATIONAL_IN_HOTEL/Turkey/rusInTurkeyHotelsCount.csv')
non = pd.read_csv('data/COUNT_NATIONAL_IN_HOTEL/Turkey/nonInTurkeyHotelsCount.csv')
bel = pd.read_csv('data/COUNT_NATIONAL_IN_HOTEL/Turkey/belInTurkeyHotelsCount.csv')
kazah = pd.read_csv('data/COUNT_NATIONAL_IN_HOTEL/Turkey/kazahInTurkeyHotelsCount.csv')
```

```python
len(ukr),len(rus),len(bel),len(kazah),len(non)
```
