```python
def getwords(doc):
    splitter = re.compile('\\W*')
    # Разбить на слова по не буквенным символам
    words = []
    for ss in splitter.split(doc):
        if 3 < len(ss) < 100:
            words.append(ss.lower())
    #Вернуть набор уникальных слов
    return words
```
